```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Charger les données depuis un fichier Excel

while True:
    Lignes = ["L12", "L15", "L17", "L48", "L05", "L06", "L08", "L47", "L50"]
    L = input("Entrer le nom de la ligne : ")
    if L in Lignes:  
        break
    else:
        print("Nom invalide. Veuillez réessayer.")

        
chemin = f"E:/ML/Journalier/Nouveau dossier/{L}.xlsx"

data = pd.read_excel(chemin)
data = data.fillna(0)

while True:

    # Demande à l'utilisateur d'entrer le nombre de produits
    while True:
        try:
            codes_number = int(input("Entrez le nombre de produits : "))
            if codes_number > 0:
                break  
            else:
                print("Le nombre de produits doit être un entier positif. Veuillez réessayer.")
        except ValueError:
            print("Entrée invalide. Veuillez entrer un nombre entier.")

    # Crée une liste vide pour stocker les codes
    codes = []

    # Demande à l'utilisateur d'entrer les noms des paramètres
    for i in range(codes_number):
        
        while True:
            C = pd.unique(data['CODE'])
            C = C.tolist()
            code = input("Entrez le code du produit {} : ".format(i+1))
            if code in C:  
                break
            else:
                print("Nom invalide. Veuillez réessayer.")
        
        codes.append(code)

    # Affiche les paramètres entrés par l'utilisateur
    print("Les codes des produits sont :", codes)



    new_data_rows = []

    for index, row in data.iterrows():
        if row['CODE'] in codes:
            new_data_rows.append(row)

    new_data = pd.DataFrame(new_data_rows)

    new_data = pd.concat([data.iloc[0:1], new_data])

    new_data.reset_index(drop=True, inplace=True)

    data = new_data


    # Demande à l'utilisateur d'entrer le nombre de paramètres d'entrée
    while True:
        try:
            features_number = int(input("Entrez le nombre de paramètres d'entrée : "))
            if features_number > 0:
                break  
            else:
                print("Le nombre doit être un entier positif. Veuillez réessayer.")
        except ValueError:
            print("Entrée invalide. Veuillez entrer un nombre entier.")
            
    # Crée une liste vide pour stocker les paramètres
    features = []

    # Demande à l'utilisateur d'entrer les noms des paramètres
    for i in range(features_number):
        
        while True:
            li = data.iloc[0]
            li = li.index.tolist()
            feature = input("Entrez le nom du paramètre d'entrée {} : ".format(i+1))
            if feature in li:  
                break
            else:
                print("Nom invalide. Veuillez réessayer.")
        
        features.append(feature)

    # Affiche les paramètres entrés par l'utilisateur
    print("Les paramètres d'entrée sont :", features)



    # Demande à l'utilisateur d'entrer le nombre de paramètres cibles
    while True:
        try:
            targets_number = int(input("Entrez le nombre de paramètres cibles : "))
            if targets_number > 0:
                break  
            else:
                print("Le nombre doit être un entier positif. Veuillez réessayer.")
        except ValueError:
            print("Entrée invalide. Veuillez entrer un nombre entier.")
            
    # Crée une liste vide pour stocker les paramètres
    targets = []

    # Demande à l'utilisateur d'entrer les noms des paramètres
    for i in range(targets_number):
        
        while True:
            li = data.iloc[0]
            li = li.index.tolist()
            target = input("Entrez le nom du paramètre cible {} : ".format(i+1))
            if target in li:  
                break
            else:
                print("Nom invalide. Veuillez réessayer.")
        
        targets.append(target)


    # Affiche les paramètres entrés par l'utilisateur
    print("Les paramètres cibles sont :", targets)


    # Séparer les caractéristiques
    features_data = data[features]
    targets_data = data[targets]

    # Normaliser les caractéristiques
    scaler = MinMaxScaler()
    features_scaled = scaler.fit_transform(features_data)

    # Créer le modèle de Deep Learning
    model = Sequential()
    model.add(Dense(10, input_dim=features_number, activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(targets_number, activation='linear'))

    # Compiler le modèle
    model.compile(loss='mean_squared_error', optimizer='adam')

    # Entraîner le modèle
    model.fit(features_scaled, targets_data, epochs=100, batch_size=32, verbose=1)

    while True:
        
        # Obtenir de nouvelles données d'entrée de l'utilisateur
        valeurs = []
        for i in range(features_number):
            while True:
                try:
                    valeur = float(input("Entrez la valeur numérique pour le paramètre d'entrée {} : ".format(features[i])))
                    break  
                except ValueError:
                        print("Entrée invalide. Veuillez entrer un nombre entier.")
            
            valeurs.append(valeur)
            print(valeur)

        new = {}
        for i in range(features_number):
            cle = f"{features[i]}"
            v = [valeurs[i]]
            new[cle] = v

        new_data = pd.DataFrame(new)

        new_data_scaled = scaler.transform(new_data)

        # Prédire les valeurs
        predictions = model.predict(new_data_scaled)

        rounded_predictions = np.round(predictions.flatten(), 2)
        print("Prédictions :")
        for i in range (targets_number):
            print(targets[i], "=", rounded_predictions[i])

        # Demander à l'utilisateur s'il souhaite effectuer une autre prédiction
        choice = input("Souhaitez-vous effectuer une autre prédiction ? (Oui/Non): ")
        if choice.lower() != "oui":
            break

    # Demander à l'utilisateur s'il souhaite changer de paramètres
    choice = input("Souhaitez-vous définir de nouveaux paramètres ? (Oui/Non): ")
    if choice.lower() != "oui":
        break
```
